```python
from sklearn.preprocessing import LabelEncoder
encoders = {}
mappings = {}
columns_to_encode = [
    'cases.0.project.project_id',
    'cases.0.disease_type',
    'cases.0.diagnoses.0.primary_diagnosis',
    'cases.0.demographic.gender',
    'cases.0.demographic.race',
    'cases.0.demographic.ethnicity',
    'cases.0.demographic.vital_status',
]
# Encode each column and store mappings
for column in columns_to_encode:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    encoders[column] = le
    mappings[column] = dict(enumerate(le.classes_))

# Print mappings for each column
for column, mapping in mappings.items():
    print(f"Mapping for {column}: {mapping}")

# Output the transformed data
print("\nTransformed Data:")
print(data)
```

```python
from sklearn.linear_model import ElasticNet
data_betta=betta.iloc[:,1:]
data_betta.index = betta.iloc[:, 0]
data_betta=data_betta.astype(float)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_betta.iloc[:,:-1],data_betta.iloc[:,-1], test_size=0.2)
from sklearn.metrics import mean_squared_error,r2_score
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.2, max_iter=10000, random_state=42)
elastic_net.fit(X_train, y_train)
y_pred = elastic_net.predict(X_test)
R2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print('Mean squared error:', mse)
print('r2_score:', R2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Actual age from your dataset (converted from days to years)
actual_age = list(data["cases.0.diagnoses.0.age_at_diagnosis"] / 365)

# Predicted age (from your regression model)
predicted_age = intercept + np.dot(filtered_betta_1, coefficients)

# Compute R^2 Score
r2 = r2_score(actual_age, predicted_age)

# Fit a regression line based on actual vs. predicted values
slope, intercept_fit = np.polyfit(actual_age, predicted_age, 1)  # Linear regression line

# Generate regression line values
fit_line = [slope * x + intercept_fit for x in actual_age]

# Plot Actual vs. Predicted DNA Methylation Age
plt.figure(figsize=(7, 5))
plt.scatter(actual_age, predicted_age, color='blue', alpha=0.6, label="Predicted vs. Actual")

# Plot Best-Fit Regression Line
plt.plot(actual_age, fit_line, color='red', linestyle='-', label="Best-Fit Regression Line")

# Reference Line (Ideal y = x)
#plt.plot([min(actual_age), max(actual_age)], [min(actual_age), max(actual_age)], 'r--', label="Ideal Fit (y=x)")

# Add R^2 value to the plot
plt.text(min(actual_age) +17, max(predicted_age) - 17, f"$R^2 = {r2:.3f}$", fontsize=12, color='black')

# Labels and Title
plt.xlabel("Actual DNA Methylation Age")
plt.ylabel("Predicted DNA Methylation Age")
plt.title("Actual vs. Predicted DNA Methylation Age")
plt.legend()
plt.grid(True)

# Show plot
plt.show()

```
